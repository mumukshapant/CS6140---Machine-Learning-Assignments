There are 5 paramaters to estimate: ùúÉ=(ùë§,ùúá1,ùúé1^2,ùúá2,ùúé2^2) where ùë§ is the probability that the data comes from the first normal probability distribution and (1-ùë§) comes from the second normal probability distribution.
The probability density function (PDF) of the mixture model is: ùëì(ùë•|ùúÉ)=ùë§ ùëì1(ùë• | ùúá1,ùúé1^2)+(1‚àíùë§) ùëì2(ùë• | ùúá2,ùúé2^2)
Your goal is to best fit a given probability density by finding ùúÉ=(ùë§,ùúá1,ùúé1^2,ùúá2,ùúé2^2) through EM iterations.

````python
import numpy as np
from math import sqrt, pi, exp
random_seed=36784765
np.random.seed(random_seed)
import matplotlib.pyplot as plt
import seaborn as sns


Mean1 = 9.0  # Input parameter, mean of first normal probability distribution
Standard_dev1 = 5.0 #@param {type:"number"}

# Mean1 = 3.5 # Input parameter, mean of second normal  probability distribution
# Standard_dev1 = 2.0 #@param {type:"number"}


Mean2 = 2.0 # Input parameter, mean of second normal  probability distribution
Standard_dev2 = 2.0 #@param {type:"number"}


# generate data
y1 = np.random.normal(Mean1, Standard_dev1, 500)
y2 = np.random.normal(Mean2, Standard_dev2, 2000)
data=np.append(y1,y2)
print("y1.shape", y1.shape)
print("y2.shape", y2.shape)
print("data.shape", data.shape)
# ```

# (1) Using a single Gaussion to estimate and draw a picure to see the result: (5 points)
# ```python
class Gaussian:
    # "Model univariate Gaussian"
    def __init__(self, mu, sigma):
        #mean and standard deviation
        
        self.mu = mu
        self.sigma = sigma


    #probability density function
    def pdf(self, datum):
        "Probability of a data point given the current parameters"

        u = (datum - self.mu) / abs(self.sigma)
        y = (1 / (sqrt(2 * pi) * abs(self.sigma))) * exp(-u * u / 2)
        return y
    
    
# estimate parameters
mu = data.mean()
sigma = data.std()

# create a Gaussian object
g = Gaussian(mu, sigma)


#==================================>

# You are correct that the combined data, which is a mixture of two Gaussians, should have two peaks. 
# However, the method you've used to estimate a single Gaussian distribution doesn't capture this. 
# When you take the mean and standard deviation of the combined data, you're effectively "smoothing"
# over the two separate distributions and creating a single "average" Gaussian that doesn't represent 
# the true structure of the data well. This is why you only see a single peak in your plot for the combined data.

# The Expectation-Maximization (EM) algorithm is designed to address this issue. It estimates the 
# parameters of multiple Gaussian distributions simultaneously, taking into account the fact that 
# the data is a mixture of these distributions. As a result, it can capture the individual peaks of each underlying distribution.

# So, in this exercise, it seems that the purpose of fitting a single Gaussian first is to show that 
# this method is insufficient for modeling the data. Then, the EM algorithm will be introduced to 
# show how it can capture the two separate Gaussians, yielding a model that fits the data much better.



# estimate parameters for y1 and y2
mu1 = y1.mean()
sigma1 = y1.std()
mu2 = y2.mean()
sigma2 = y2.std()

# create Gaussian objects
g1 = Gaussian(mu1, sigma1)
g2 = Gaussian(mu2, sigma2)

# calculate PDF for a range of x values
x = np.linspace(min(data), max(data), num=1000)
pdf1 = np.array([g1.pdf(i) for i in x])
pdf2 = np.array([g2.pdf(i) for i in x])

# plot
sns.distplot(y1, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.plot(x, pdf1, color='red')
plt.title('Estimated Gaussian Distribution for y1')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()

sns.distplot(y2, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.plot(x, pdf2, color='red')
plt.title('Estimated Gaussian Distribution for y2')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()


#=====================================>

# calculate PDF for a range of x values
x = np.linspace(min(data), max(data), num=1000)
pdf = np.array([g.pdf(i) for i in x])

# plot
# sns.distplot(data, hist=False, kde=True, 
#              color = 'darkblue', 
#              hist_kws={'edgecolor':'black'},
#              kde_kws={'linewidth': 4})

plt.plot(x, pdf, color='red')
plt.title('Estimated Gaussian Distribution')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()




````

````python
# (2) Using a 2 Gaussian mixture model to estimate and draw a picture to see the result(Do not use sklearn GaussianMixture): (30 points)
# ```python

from random import uniform
from math import sqrt, log, exp, pi

class GaussianMixture_self:
# "Model mixture of two univariate Gaussians and their EM estimation"

    def __init__(self, data, mu_min=min(data), mu_max=max(data), sigma_min=1, sigma_max=1, mix=.5):
        
        self.data = data
        self.gaussian_one = Gaussian(uniform(mu_min, mu_max),
                                     uniform(sigma_min, sigma_max))
        self.gaussian_two = Gaussian(uniform(mu_min, mu_max),
                                     uniform(sigma_min, sigma_max))
        
        self.mix = mix

    def Estep(self):
        "Perform an E(stimation)-step, assign each point to gaussian 1 or 2 with a percentage"
        self.loglikelihood = 0
        
        for datapoint in self.data:
            
            
            responsibility_1 = self.gaussian_one.pdf(datapoint) * self.mix
            responsibility_2 = self.gaussian_two.pdf(datapoint) * (1-self.mix)
            
            # sum two responsibilities.. (denominator of Bayes Theorem)
            denominator = responsibility_1 + responsibility_2
            
            # E-step, normalize the responsibilities. 
            responsibility_1 /= denominator
            responsibility_2 /= denominator
            
            # update the log-likelihood with the log of the sum of normalized probabilities
            self.loglikelihood += log(responsibility_1+responsibility_2)
            
            # return a generator that outputs the pair of normalzied weights (responsbilities) for the current data poitn. 
            # this will be used in the M-step to update the Gaussian distributions. 
            yield (responsibility_1,responsibility_2)
            
            
            
            
        

    def Mstep(self, weights):
        "Perform an M(aximization)-step"
        
        # compute denominators
        # unpacks the responsibilities of the two gaussians for each data point, collected from E-step
        (responsibility_1,responsibility_2) = zip(*weights)
        
        # sum up the responsibilities of the first and second gaussians across all datapoints. 
        # these sums will be used as denominators to calculate the updates parameters of the Gaussians.
        one_resp1_denom = sum(responsibility_1)
        two_resp2_denom = sum(responsibility_2)
        # compute new means
        
        # compute the update means of the two gaussians. for each gaussian, the update mean is the weighted sum of the data pointns, where the weights are the responsibilities of that
        # gaussian for each data point. The sums are then normalized by the total responsibilities of the respective gaussians
        
        self.gaussian_one.mu = sum(w * d / one_resp1_denom for (w, d) in zip(responsibility_1, self.data))
        self.gaussian_two.mu = sum(w * d / two_resp2_denom for (w, d) in zip(responsibility_2, self.data))
        
        
        # compute the updated standard deviations
        self.gaussian_one.sigma = sqrt(sum(w * ((d - self.gaussian_one.mu) ** 2)
                                  for (w, d) in zip(responsibility_1, self.data)) / one_resp1_denom)
        self.gaussian_two.sigma = sqrt(sum(w * ((d - self.gaussian_two.mu) ** 2)
                                  for (w, d) in zip(responsibility_2, self.data)) / two_resp2_denom)
        
        # finallly mixing weight is updated for first gaussian
        self.mix = one_resp1_denom / len(self.data)




    def iterate(self, N=1, verbose=False):
        "Perform N iterations, then compute log-likelihood"
        try:
            for i in range(1, N+1):
                self.Mstep(self.Estep())
                if verbose:
                    print('{0:2} {1}'.format(i, self))
        except StopIteration:
            pass
        return self
    

    def pdf(self, x):
        return (self.mix)*self.gaussian_one.pdf(x) + (1-self.mix)*self.gaussian_two.pdf(x)
        
mix = GaussianMixture_self(data)
mix.iterate(N=50)
mixture_pdf = [mix.pdf(x) for x in np.linspace(min(data), max(data), num=1000)]

import matplotlib.pyplot as plt

x_values = np.linspace(min(data), max(data), num=1000)
plt.plot(x_values, mixture_pdf)
plt.show()
````
