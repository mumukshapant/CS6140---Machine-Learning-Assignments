## Part A: Show the Outliers and the box plot

````python
# Calculation of outliers

import numpy as np
import matplotlib.pyplot as plt

data = [11, 3, 8, 10, 12, 5, 1, 50]
# Sort the data
data.sort()

# Calculate Q1, Q2(Q median), Q3
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

# Calculate IQR
IQR = Q3 - Q1

# Determine the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Create a box plot
ax.boxplot(data, vert=False, patch_artist=True, notch=False, sym='r.')

# Set x-axis label
ax.set_xlabel('Data Values')

# Set title
ax.set_title('Box plot with Outliers')

# Show the plot
plt.show()
# This code will both calculate and print the outliers and create a box plot highlighting the outliers.


````

````python
# Specify four different thresholds and corresponding colors
thresholds = [3, 3.5,4,4.5]
colors = ['red', 'green', 'blue', 'purple']

# Calculate mean and standard deviation for 'SalePrice' column
mean_saleprice = df['SalePrice'].mean()
std_dev_saleprice = df['SalePrice'].std()

# Calculate z-scores for 'SalePrice' column
z_scores_saleprice = (df['SalePrice'] - mean_saleprice) / std_dev_saleprice

# Identify and plot outliers for each threshold
for threshold, color in zip(thresholds, colors):
    # Identify outliers
    outliers = df['SalePrice'][np.abs(z_scores_saleprice) > threshold]
    print(f"Outliers for threshold {threshold}: {outliers}")
    print(f"Number of outliers removed for threshold {threshold}: {len(outliers)}")
    # Create a new figure
    plt.figure(figsize=(10, 6))
    
    # Plot all data points
    plt.scatter(df.index, df['SalePrice'], color='gray', label='Data points')
    
    # Plot outliers
    plt.scatter(outliers.index, outliers, color=color, label=f'Outliers (|z| > {threshold})')
    
    # Set x and y labels
    plt.xlabel('Index')
    plt.ylabel('SalePrice')
    
    # Set title
    plt.title(f'SalePrice Outliers with Z-Score Threshold = {threshold}')
    
    # Add legend
    plt.legend()
    
    # Show plot
    plt.show()
````

````python
from sklearn.neighbors import LocalOutlierFactor

# Extract 'SalePrice' and 'LotArea' columns for outlier detection
dataset = df[['SalePrice', 'LotArea']].values

# Initialize the LocalOutlierFactor model
local_outlier_factor = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

# Use fit_predict to compute the predicted labels of the data points
labels_predicted = local_outlier_factor.fit_predict(dataset)
outlier_scores = local_outlier_factor.negative_outlier_factor_

# Count the number of outliers
num_outliers = (labels_predicted == -1).sum()

plt.title("Local Outlier Factor (LOF)")
plt.scatter(dataset[:, 0], dataset[:, 1], color="k", s=3.0, label="Data points")
# Plot circles with radius proportional to the outlier scores
radius = (outlier_scores.max() - outlier_scores) / (outlier_scores.max() - outlier_scores.min())
plt.scatter(
    dataset[:, 0],
    dataset[:, 1],
    s=1000 * radius,
    edgecolors="r",
    facecolors="none",
    label="Outlier scores",
)
plt.axis("tight")
plt.xlabel("SalePrice")
plt.ylabel("LotArea")
plt.xlim((df['SalePrice'].min(), df['SalePrice'].max()))
plt.ylim((df['LotArea'].min(), df['LotArea'].max()))
plt.title(f"Outliers detected: {num_outliers}")
legend = plt.legend(loc="upper left")
legend.legendHandles[0]._sizes = [10]
legend.legendHandles[1]._sizes = [20]
plt.show()
````

## 2. PCA (35 points)

Accuracy Comparison for Logistic Regression model: before and after PCA.
Please follow the following steps:

  1. Seperate and standardize the disease classification dataset. (5 points)
  2. Do Eigen decomposition using any LA library of your choice. Display scree plot. (10 points)
  3. Primary Component Selection. (Select the first 6 components) (5 points)
  4. Projection in a New Feature Space. (5 points)
  5. Principal Component Analysis. (5 points)
  6. Compare the presision and recall for the data using logistic regression before and after PCA. (10 points)

## Part 1

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Load the data
dataframe = pd.read_csv("Q2_dataset.csv")

# Separate features (features_df) and target (target_df)
features_df = dataframe.drop('class', axis=1)  # Features (input variables)
target_df = dataframe['class']               # Target variable (labels)

# Assuming 'features_df' contains numerical features that need to be standardized
scaler_object = StandardScaler()
standardized_features = scaler_object.fit_transform(features_df)

# Convert the standardized features back to a DataFrame 
standardized_features_df = pd.DataFrame(standardized_features, columns=features_df.columns)

# Display the first few rows of the standardized DataFrame
print(standardized_features_df.head())
````

## Part 2

````python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Perform Eigen decomposition
eigen_values, eigen_vectors = np.linalg.eigh(standardized_features_df.cov())

# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:, i]) for i in range(len(eigen_values))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs.sort(key=lambda x: x[0], reverse=True)

# Separate the eigenvalues and eigenvectors into two lists
eigen_values_sorted = [eigen_pairs[i][0] for i in range(len(eigen_pairs))]
eigen_vectors_sorted = [eigen_pairs[i][1] for i in range(len(eigen_pairs))]

# Calculate the explained variance ratio for each eigenvalue
explained_variance_ratio = eigen_values_sorted / np.sum(eigen_values_sorted)

# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, len(eigen_values_sorted) + 1), explained_variance_ratio, marker='o', linestyle='-')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
````

## Part 3, 4, 5

````python
# Select the first 6 principal components
selected_eigenvectors = np.array([eigen_vectors_sorted[i] for i in range(6)]).T

# Perform projection in the new feature space
projected_data = standardized_features_df.values.dot(selected_eigenvectors)

print(projected_data)
````

## Part 6

````python
# Original data: Splitting into training and testing sets
X_train_beforePCA, X_test_beforePCA, y_train, y_test = train_test_split(features_df, target_df, test_size=0.3, random_state=42)

# Original data: Training a logistic regression model
logistic_regression_beforePCA = LogisticRegression()
logistic_regression_beforePCA.fit(X_train_beforePCA, y_train)

# Original data: Evaluating model's performance with accuracy, precision, and recall
y_pred_beforePCA = logistic_regression_beforePCA.predict(X_test_beforePCA)
accuracy_beforePCA = accuracy_score(y_test, y_pred_beforePCA)
precision_beforePCA = precision_score(y_test, y_pred_beforePCA)
recall_beforePCA = recall_score(y_test, y_pred_beforePCA)

# Display performance metrics for logistic regression on original data
print("Logistic Regression Performance Before PCA:")
print("Accuracy:", accuracy_beforePCA)
print("Precision:", precision_beforePCA)
print("Recall:", recall_beforePCA)

# Transformed data (after PCA): Splitting into training and testing sets
X_train_afterPCA, X_test_afterPCA, _, _ = train_test_split(projected_data, target_df, test_size=0.3, random_state=42)

# Transformed data (after PCA): Training a logistic regression model
logistic_regression_afterPCA = LogisticRegression()
logistic_regression_afterPCA.fit(X_train_afterPCA, y_train)

# Transformed data (after PCA): Evaluating model's performance with accuracy, precision, and recall
y_pred_afterPCA = logistic_regression_afterPCA.predict(X_test_afterPCA)
accuracy_afterPCA = accuracy_score(y_test, y_pred_afterPCA)
precision_afterPCA = precision_score(y_test, y_pred_afterPCA)
recall_afterPCA = recall_score(y_test, y_pred_afterPCA)

# Display performance metrics for logistic regression on transformed data (after PCA)
print("\nLogistic Regression Performance After PCA:")
print("Accuracy:", accuracy_afterPCA)
print("Precision:", precision_afterPCA)
print("Recall:", recall_afterPCA)
````

## 3. EM Algorithm (35 points)

Etimate the probability distribution in a 1-dimensional dataset
There are two Normal distributions  𝑁(𝜇1,𝜎1^2) and 𝑁(𝜇2,𝜎2^2).
There are 5 paramaters to estimate: 𝜃=(𝑤,𝜇1,𝜎1^2,𝜇2,𝜎2^2) where 𝑤 is the probability that the data comes from the first normal probability distribution and (1-𝑤) comes from the second normal probability distribution.
The probability density function (PDF) of the mixture model is: 𝑓(𝑥|𝜃)=𝑤 𝑓1(𝑥 | 𝜇1,𝜎1^2)+(1−𝑤) 𝑓2(𝑥 | 𝜇2,𝜎2^2)
Your goal is to best fit a given probability density by finding 𝜃=(𝑤,𝜇1,𝜎1^2,𝜇2,𝜎2^2) through EM iterations.

````python
import numpy as np
from math import sqrt, pi, exp
random_seed=36784765
np.random.seed(random_seed)
import matplotlib.pyplot as plt
import seaborn as sns


Mean1 = 9.0  # Input parameter, mean of first normal probability distribution
Standard_dev1 = 5.0 #@param {type:"number"}

# Mean1 = 3.5 # Input parameter, mean of second normal  probability distribution
# Standard_dev1 = 2.0 #@param {type:"number"}


Mean2 = 2.0 # Input parameter, mean of second normal  probability distribution
Standard_dev2 = 2.0 #@param {type:"number"}


# generate data
y1 = np.random.normal(Mean1, Standard_dev1, 500)
y2 = np.random.normal(Mean2, Standard_dev2, 2000)
data=np.append(y1,y2)
print("y1.shape", y1.shape)
print("y2.shape", y2.shape)
print("data.shape", data.shape)
# ```

# (1) Using a single Gaussion to estimate and draw a picure to see the result: (5 points)
# ```python
class Gaussian:
    # "Model univariate Gaussian"
    def __init__(self, mu, sigma):
        #mean and standard deviation
        
        self.mu = mu
        self.sigma = sigma


    #probability density function
    def pdf(self, datum):
        "Probability of a data point given the current parameters"

        u = (datum - self.mu) / abs(self.sigma)
        y = (1 / (sqrt(2 * pi) * abs(self.sigma))) * exp(-u * u / 2)
        return y
    
    
# estimate parameters
mu = data.mean()
sigma = data.std()

# create a Gaussian object
g = Gaussian(mu, sigma)


#==================================>

# You are correct that the combined data, which is a mixture of two Gaussians, should have two peaks. 
# However, the method you've used to estimate a single Gaussian distribution doesn't capture this. 
# When you take the mean and standard deviation of the combined data, you're effectively "smoothing"
# over the two separate distributions and creating a single "average" Gaussian that doesn't represent 
# the true structure of the data well. This is why you only see a single peak in your plot for the combined data.

# The Expectation-Maximization (EM) algorithm is designed to address this issue. It estimates the 
# parameters of multiple Gaussian distributions simultaneously, taking into account the fact that 
# the data is a mixture of these distributions. As a result, it can capture the individual peaks of each underlying distribution.

# So, in this exercise, it seems that the purpose of fitting a single Gaussian first is to show that 
# this method is insufficient for modeling the data. Then, the EM algorithm will be introduced to 
# show how it can capture the two separate Gaussians, yielding a model that fits the data much better.



# estimate parameters for y1 and y2
mu1 = y1.mean()
sigma1 = y1.std()
mu2 = y2.mean()
sigma2 = y2.std()

# create Gaussian objects
g1 = Gaussian(mu1, sigma1)
g2 = Gaussian(mu2, sigma2)

# calculate PDF for a range of x values
x = np.linspace(min(data), max(data), num=1000)
pdf1 = np.array([g1.pdf(i) for i in x])
pdf2 = np.array([g2.pdf(i) for i in x])

# plot
sns.distplot(y1, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.plot(x, pdf1, color='red')
plt.title('Estimated Gaussian Distribution for y1')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()

sns.distplot(y2, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.plot(x, pdf2, color='red')
plt.title('Estimated Gaussian Distribution for y2')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()


#=====================================>

# calculate PDF for a range of x values
x = np.linspace(min(data), max(data), num=1000)
pdf = np.array([g.pdf(i) for i in x])

# plot
# sns.distplot(data, hist=False, kde=True, 
#              color = 'darkblue', 
#              hist_kws={'edgecolor':'black'},
#              kde_kws={'linewidth': 4})

plt.plot(x, pdf, color='red')
plt.title('Estimated Gaussian Distribution')
plt.xlabel('Data'); plt.ylabel('Probability Density');
plt.show()




````

````python
# (2) Using a 2 Gaussian mixture model to estimate and draw a picture to see the result(Do not use sklearn GaussianMixture): (30 points)
# ```python

from random import uniform
from math import sqrt, log, exp, pi

class GaussianMixture_self:
# "Model mixture of two univariate Gaussians and their EM estimation"

    def __init__(self, data, mu_min=min(data), mu_max=max(data), sigma_min=1, sigma_max=1, mix=.5):
        
        self.data = data
        self.gaussian_one = Gaussian(uniform(mu_min, mu_max),
                                     uniform(sigma_min, sigma_max))
        self.gaussian_two = Gaussian(uniform(mu_min, mu_max),
                                     uniform(sigma_min, sigma_max))
        
        self.mix = mix

    def Estep(self):
        "Perform an E(stimation)-step, assign each point to gaussian 1 or 2 with a percentage"
        self.loglikelihood = 0
        
        for datapoint in self.data:
            
            
            responsibility_1 = self.gaussian_one.pdf(datapoint) * self.mix
            responsibility_2 = self.gaussian_two.pdf(datapoint) * (1-self.mix)
            
            # sum two responsibilities.. (denominator of Bayes Theorem)
            denominator = responsibility_1 + responsibility_2
            
            # E-step, normalize the responsibilities. 
            responsibility_1 /= denominator
            responsibility_2 /= denominator
            
            # update the log-likelihood with the log of the sum of normalized probabilities
            self.loglikelihood += log(responsibility_1+responsibility_2)
            
            # return a generator that outputs the pair of normalzied weights (responsbilities) for the current data poitn. 
            # this will be used in the M-step to update the Gaussian distributions. 
            yield (responsibility_1,responsibility_2)
            
            
            
            
        

    def Mstep(self, weights):
        "Perform an M(aximization)-step"
        
        # compute denominators
        # unpacks the responsibilities of the two gaussians for each data point, collected from E-step
        (responsibility_1,responsibility_2) = zip(*weights)
        
        # sum up the responsibilities of the first and second gaussians across all datapoints. 
        # these sums will be used as denominators to calculate the updates parameters of the Gaussians.
        one_resp1_denom = sum(responsibility_1)
        two_resp2_denom = sum(responsibility_2)
        # compute new means
        
        # compute the update means of the two gaussians. for each gaussian, the update mean is the weighted sum of the data pointns, where the weights are the responsibilities of that
        # gaussian for each data point. The sums are then normalized by the total responsibilities of the respective gaussians
        
        self.gaussian_one.mu = sum(w * d / one_resp1_denom for (w, d) in zip(responsibility_1, self.data))
        self.gaussian_two.mu = sum(w * d / two_resp2_denom for (w, d) in zip(responsibility_2, self.data))
        
        
        # compute the updated standard deviations
        self.gaussian_one.sigma = sqrt(sum(w * ((d - self.gaussian_one.mu) ** 2)
                                  for (w, d) in zip(responsibility_1, self.data)) / one_resp1_denom)
        self.gaussian_two.sigma = sqrt(sum(w * ((d - self.gaussian_two.mu) ** 2)
                                  for (w, d) in zip(responsibility_2, self.data)) / two_resp2_denom)
        
        # finallly mixing weight is updated for first gaussian
        self.mix = one_resp1_denom / len(self.data)




    def iterate(self, N=1, verbose=False):
        "Perform N iterations, then compute log-likelihood"
        try:
            for i in range(1, N+1):
                self.Mstep(self.Estep())
                if verbose:
                    print('{0:2} {1}'.format(i, self))
        except StopIteration:
            pass
        return self
    

    def pdf(self, x):
        return (self.mix)*self.gaussian_one.pdf(x) + (1-self.mix)*self.gaussian_two.pdf(x)
        
mix = GaussianMixture_self(data)
mix.iterate(N=50)
mixture_pdf = [mix.pdf(x) for x in np.linspace(min(data), max(data), num=1000)]

import matplotlib.pyplot as plt

x_values = np.linspace(min(data), max(data), num=1000)
plt.plot(x_values, mixture_pdf)
plt.show()
````
